```python
import pandas as pd
path = '../data/raw/steam_games.csv'
data = pd.read_csv(path)
print(data.shape)
data.head()
```

```python
data = data.dropna()
columns_to_process = ['developers', 'publishers', 'categories', 'genres', 'platforms']
unwanted_values = {'', '(none)', '(NONE)', '(None)'} 

for col in columns_to_process:
    data[col] = data[col].apply(
        lambda x: [item.strip() for item in eval(x) if item.strip() and item.strip().lower() not in unwanted_values] 
        if isinstance(x, str) else 
        [item.strip() for item in x if item.strip() and item.strip().lower() not in unwanted_values]
    )

data = data[data['is_released'] == True]

data = data.drop(columns=['is_released'])

data['release_date'] = data['release_date'].astype(str)
data['release_date'] = data['release_date'].str.replace(" 00:00:00", "", regex=False)
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce', format='%Y-%m-%d')

data.head()
```

```python
data.to_csv("../data/processed/cleaned_games.csv") 
```
